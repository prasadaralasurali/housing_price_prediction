## Import Data
The dataset used here originates from the [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Housing). The Boston housing data was collected in 1978 and each of the 506 entries represent aggregated data about 14 features for homes from various suburbs in Boston, Massachusetts. The following preprocessing steps have been made to the dataset:
- The features `'RM'`, `'LSTAT'`, `'PTRATIO'`, and `'MEDV'` are essential (see below for description of these variables). The remaining non-relevant features have been excluded.
- 16 data points have an `'MEDV'` value of 50.0. These data points likely contain **missing or censored values** and have been removed.
- 1 data point has an `'RM'` value of 8.78. This data point can be considered an **outlier** and has been removed.

**Features**
1.  `RM`: average number of rooms per dwelling
2. `LSTAT`: percentage of population considered lower status

**Target Variable**
4. `MEDV`: median value of owner-occupied homes

### Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. I have used [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination) (R<sup>2</sup>) to quantify my model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what fraction of variation in the target variable can be explained by the **features** using this model. _A model can be given a negative R<sup>2</sup> as well, which indicates that the model is arbitrarily worse than one that always predicts the mean of the target variable._

The `performance_metric` function in the code cell below returns R<sup>2</sup> between 'y_true' and 'y_predict'

```python
# Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

### Shuffle and Split Data
I have split the data into training and testing subsets. The data has been shuffled into a random order before creating the training and testing subsets to remove any bias in the ordering of the dataset.

One of the critical steps in machine learning is to evaluate how well a model that we build performs. Training a model and testing it on the same data is a big mistake in machine learning. By adding high enough complexity to a model (e.g., high enough depth in decision tree or high degree of polynomials in polynomial regression) it is possible to achieve very high score if we evaluate the model on the same data that we used to built it. But such models are unlikely to perform well on unseen data. Because model has just memorized the data instead of finding general pattern. If we split the data into training and testing set, and trian the model on training set and test it on test set, we can easily figure out which model overfits and which model is likely to generalize well on unseen data. However, it should be noted that train_test_split alone does not guarantee accurate evaluation of predictive perrformance of a model. For instance, if the sample from which you collected your dataset is biased and not representative of the underlying population, even the model that performed well on test set is unlikely to perform well on new observations (ones that are not part of the dataset). 


In the code cell below, I have implemented the following:
- Used `train_test_split` from `sklearn.cross_validation` to shuffle and split the `features` and `prices` data into training and testing sets at a 80:20% ration.
- Set the `random_state` for `train_test_split` to ensure results are consistent.

```python
# Import 'train_test_split'
from sklearn.model_selection import train_test_split

# Shuffle and split the data into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size = 0.2, random_state = 42)
# by default shuffle = True in train_test_split function
```

### Bias–Variance Tradeoff 
The model trained with a maximum depth of 1 suffers from high bias, wherease the model trained with a maximum depth of 10 suffers from high variance.

At max_depth = 1, both training score and validation score are low, which is a sign of underfitting. When max_depth = 10, the training score is very high but validation score is low, a sign of overfitting. Upto a certain point, as model complexity increased, validation score increased, but afterwards due to overfitting it started decreasing. Training score kept increasing, as model complexity increased.

**In summary,**

max_depth = 1 -> simple model -> high bias (low variance) -> underfitting    
max_depth = 10 -> too complex model -> high variance (low bias) -> overfitting   

As we increase maximum deapth from 1 to 3, validation score keeps increasing becuase of decreasing *error due to bias*. However, increase of max depth beyond 4 reduces validation score because of increasing *error due to variance*. Therefore maximum depth of 3 provides a good balance of bias and variance, and hence likely to generalizes well to unseen data.  

Although, max_depth of 3 and 4 have similar validation score, going by Occams Razor, I have chosen max_depth = 3. It should be noted that the model with max_depth = 4 has a slightly higher validation score, but I think the difference is very negligible to be of practical significance.

### Grid Search

Grid search is an optimization technique to find out the best combination of hyper-parameters for a machine learning algorithm that produces optimal results. For instance, decision tree has hyper-parameters like type of kernel used, maximum depth of tree, min_samples_split etc. Grid search builds multiple models using different combinations of user-defined hyper parameters (each grid representing specific combination of hyper-parameters), evaluate these models using a user-defined evaluation metrics (or a default metric if user has not defined a scoring metric), and outputs the best combination of hyperparameters that will produce the highest model performance with repsect to evaluation metric used.

---
<a id='step11'></a>
### K-Fold Cross-Validation
K-fold cross-validation (CV) divides the the training set into k folds of approximately equal size. In the first iteration, the first fold is treated as a validation set, and the model is fit on the remaining k − 1 folds. The error is then computed on the the validation set. This procedure is repeated k times and each time a different fold is treated as a validation set resulting in k estimates of the test errors. The final error estimate is computed by averaging these k estimates (Reference 1).

When evaluating different hyperparameters in grid search CV, even though dataset is divided into training and test set, there is still a risk of overfitting because the parameters can be tweaked until the model performs optimally on test set. This way model can learn test set as well, defeating the very purpose of having a different test set. A simple solution to this problem would be to divide data into three sets; training, validation and test set. Training is done on the training set, evaluation is done on the validation set, and final evaluation of the chosen model can be done on the test set to report its performance. However, a drawback to this mehtod is that number of data points available for training the model is drastically reduced which affects the model training. Furthermore, the results can depend on a particular random choice for the pair of training and validation sets. K-fold CV circumvent this problem, wherein, a test set is still held out for final evaluation, but a separate validation set is no longer needed as in different iteration of training, different subset of training set becomes validation set (Reference 2).

### Fitting a Model
In this section, I have bought everything together and trained a model using a **decision tree algorithm**, a supervised learning algorithm. I used the grid search technique to optimize the `'max_depth'` parameter for the decision tree. The `'max_depth'` parameter can be thought of as how many questions the decision tree algorithm is allowed to ask about the data before making a prediction. 

I have used `ShuffleSplit()` as an alternative form of cross-validation. While it is not the K-Fold cross-validation technique described above, this type of cross-validation technique is just as useful!. The `ShuffleSplit()` implementation below created 10 (`'n_splits'`) shuffled sets, and for each shuffle, 20% (`'test_size'`) of the data will be used as the *validation set*. 

In the `fit_model` function in the code cell below, I have implemented the following:
- Used [`DecisionTreeRegressor`](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html) from `sklearn.tree` to create a decision tree regressor object.
- Created a dictionary for `'max_depth'` with the values from 1 to 10.
- Used [`make_scorer`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html) from `sklearn.metrics` to create a scoring function object.
  - Passed the `performance_metric` function as a parameter to the object.
- Used [`GridSearchCV`](http://scikit-learn.org/0.17/modules/generated/sklearn.grid_search.GridSearchCV.html) from `sklearn.grid_search` to create a grid search object.
  - Passed the variables `'regressor'`, `'params'`, `'scoring_fnc'`, and `'cv_sets'` as parameters to the object. 

```python
# Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    # sklearn version 0.18: ShuffleSplit(n_splits=10, test_size=0.1, train_size=None, random_state=None)
    # sklearn versiin 0.17: ShuffleSplit(n, n_iter=10, test_size=0.1, train_size=None, random_state=None)
    cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.20, random_state = 0)

    # Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

    # Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth': np.arange(1, 11)}

    # Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # Create the grid search cv object 

    grid = GridSearchCV(estimator = regressor, param_grid = params, scoring = scoring_fnc, cv = cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```

### Making Predictions
Once a model has been trained on a given set of data, it can now be used to make predictions on new sets of input data. In the case of a *decision tree regressor*, the model has learned *what the best questions to ask about the input data are*, and can respond with a prediction for the **target variable**. We can use these predictions to gain information about data where the value of the target variable is unknown — such as data the model was not trained on.

#### Optimal Model

In the code block below, I have fit the decision tree regressor to the training data by using the fit_model function.

```python
# Fit the training data to the model using the fit_model function
reg = fit_model(X_train, y_train)

# Produce the value for 'max_depth'

print("Parameter 'max_depth' is {} for the optimal model.".format(reg.get_params()['max_depth']))
```

Maximum depth for the optimal model is 4 as per GridSearchCV. This compares fairly well with my guess based on complexity curve, where I chose a maximum depth of 3 among the models having max_depth of 1 to 10. I explained above why I chose max_depth = 3 although max_depth = 4 produced slightly higher vaidation score when I investigated complexity curve.

#### Model Evaluation

```python
#model evaluation
predictions = reg.predict(X_test)
print ("The model has an R-squared (coefficient of determination) of {:.2f}". format(performance_metric(y_test, predictions)))
```

### Sensitivity
An optimal model is not necessarily a robust model. Sometimes, a model is either too complex or too simple to sufficiently generalize to new data. Sometimes, a model could use a learning algorithm that is not appropriate for the structure of the data given. Other times, the data itself could be too noisy or contain too few samples to allow a model to adequately capture the target variable — i.e., the model is underfitted.

## Conclusion
Here, I have built a predictive model using a Decision Tree algorithm to predict housing price in Boston. This model can be used in real world setting, although with a great caution. While this model can be used to get a general price trend in a neighbouhood, it may not be accurate in estimating the price of a particular house, as home price depends on so many variables other than the three variables used to build this model.

Since the feature 'MEDV' has been multiplicatively scaled to account for 35 years of market inflation, the price is likely to reflect current trend. However, we need to take into account whether the features used to build the model is relavant in the same way as was 40 years ago. For instance, today people may not be willing to pay higher price just because the neighborhood has a good school to the same degree as they did in 1978. With improvement in mass transportaion, and everyone having a car these days, how far the good school is may not be equally relavant as it used to be.

The three features used here are not sufficeint to describe a home fully. The home price depends on many more factors than just three used in this model. For instance, how green the area is? Is it well connected with the public transportaion like subway/light rail? Is it close to beach/river trail? is it close to business districts? What is the crime rate in the neighbor hood? All these questions matter in determinining the price of a house. I believe, some of the 11 features in the data set that are not used in this model may be useful in improving the model accuracy.

The model does not seem to be robust, as the range of prices it predicted in multiple trials is around 69,000. However, as I mentioned above, if we are interested in estimating a general price trend, this level of robustness should be okay. 

This model is applicable only to Boston, or the cities that are similar to Boston in all respects. Applying it to some rural cities is most likely to result in inaccurate predictions. What is relevant in Boston may not be relavent to the same degree or may not be relavant at all in other rural cities.

Finally, it is not logical to judge the price of an individual home just based on the characteristics of the neighborhood it is in. The price of an individual house depend on many factors such as what kind of material is used for construction, how much area is avialable for lawn or backyard, etc. It is difficult to collect data about all those features and include them in model building. The prediction of a model, such as the one built here, helps to get a general trend, but accurate estimation of the price of an individual house needs further consideration of many other factors.
